# Setup F1 dataset for learning pyspark and delta lake

## Use the the most complete F1 dataset you can find on Kaggle.com

E.g. https://www.kaggle.com/datasets/rohanrao/formula-1-world-championship-1950-2020

## Create project folder to hold all files for learning pyspark and delta lake

If you have a project folder where your virtual environment is install add another folder called "jupyter-pyspark"

Within the new folder add a F1 source files folder called "f1-sourcefiles".

Also, add a folder called "delta" to put all delta tables created by pyspark

Create your jupyter notebooks in the "jupyter-pyspark" folder created above.
